```python
import pandas as pd
```

```python
results = pd.read_csv('Final_Results2.csv')
```

```python
results=results.set_value(4,'District','Ampara')
```

```python
import math
```

```python
GR_v = 8931+6099+14836+9817
SP_v=54150+54676+64544+44072
AKD_v = 489+18+1578+385
valid=63571+60794+80959+54275
regis = 89538+89403+117333+75383
maj = 45219+48577+49708+34255
excess = 22854+24298+25643+17320
maj=-maj
excess=-excess
gr_p=round(GR_v*100/valid,2)
sp_p=round(SP_v*100/valid,2)
akd_p=round(AKD_v*100/valid,2)
turn=round(valid*100/regis,2)
results.loc[25] = [25,'Vanni',GR_v,gr_p,SP_v,sp_p,AKD_v,akd_p,turn,valid,regis,maj,excess]
```

```python
joindf = states.merge(results, left_on='electoralDistrict', right_on='District')
```

```python
final_df=joindf.drop(['Unnamed: 0','District',], axis=1)
```

```python
final_df.rename(columns={'GR_%': 'GR_P', 'SP_%': 'SP_P','AKD_%':'AKD_P'}, inplace=True)
```

```python
final_df.to_file('electoral_with_results.geojson', driver='GeoJSON')
```

```python
resutls=[]
fstring=""
i=1
for res in array_res:
    resdic = res.split("-")
    #print(resdic)
    resi={'name':resdic[0],'vote':int(resdic[1].replace(',','').strip()),'percent':100*int(resdic[1].replace(',','').strip())/valid}
    resutls.append(resi)
    print("%d. %s %s  ( %4.2f %%)"%(i,resdic[0],resdic[1],resi['percent']))
    i+=1
```
